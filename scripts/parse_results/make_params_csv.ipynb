```python
param_dir = base_dir / "params"

import re
matcher = re.compile(r"letting (\w+) be (.*)")

names = []
def process(p):
    def transform(l):
        return matcher.findall(l.strip())[0]
    
    lines = list(p.open().readlines())
    result = [ transform(l) for l in lines[1:] ] 
    global names
    (names, vals) = list(zip(*result))
    names += ("param",)
    return vals + (p,)

param_data = [ process(p) for p in param_dir.iterdir()]
param_data
```

```python
param_df= df.from_items([ (k,v) for (k,v) in  zip(names, zip(*param_data)) ],
              columns=["n_upper",	"n_boats",	"n_periods",	"capacity",	"crew","param"])
param_df['param']= param_df['param'].apply(lambda p: p.stem );
```

```python
merged_df = df.merge(param_df)
cols=[
 'n_upper',
 'n_boats',
 'n_periods',
 'capacity',
 'crew',
 'ordering',
 'quality',
 'eprimes_count',
 'Satisfiable',
 'MaxSolutions',
 'eprimes',
 'param']

merged_df.to_csv( str(base_dir / 'param_data_all.csv'), cols=cols )
with (base_dir / 'param_data_all.html').open("w") as f:
    f.writelines(merged_df[cols].to_html())
    
df.to_html(
```
