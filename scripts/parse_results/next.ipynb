```python
def plot_grid(*, our, cols_names, use_discriminating=False, legend=True, colors=None ):
    
    if use_discriminating:
        values='discriminating'
        x_axis_name='number of distinguishing params'
        pass
    else:
        values='resulting_models_filled'
        x_axis_name='models left'
    
    pt=our.pivot_table(rows=['total_timeout_h'],
                       cols=cols_names,
                       values=values)
    
    rows = our[cols_names[0]].unique()
    cols = our[cols_names[1]].unique()
    print("rows:", rows, "cols:", cols)
    
    r_limits =( our['total_timeout_h'].dropna().min(),  our['total_timeout_h'].dropna().max())
    
    lens = ( (i,j, pt[r][c].columns ) for i, r in  enumerate(rows)  for j, c in enumerate(cols) )
    max_val = max(lens,key=lambda v: len(v[2]))
    *max_val, _ = max_val
    
    legend_names=our[cols_names[2]].unique()
    if colors:
        styles = colors
    else:
        styles = ['#0000FF', '#FF00FF', '#008080', '#008000', '#FFFF00', '#00FF00', '#800080', '#00FFFF', '#C0C0C0', '#000080', 
                  '#FF0000', '#808000', '#000000', '#FFA500', '#808080', '#800000']
    display(legend_names)
    styles_mapping = make_style_mapping(legend_names, styles)
    display( [ ele for ele in sorted(styles_mapping.items()) if len(ele[0]) == 1  ] )
    #     display( [ ele for ele in sorted(styles_mapping.keys())
#               if set(['ksample_first-100.0-10.0', 'ksample_first-20.0-10.0', 'ksample_halving-100.0-10.0',
#                       'ksample_halving-20.0-10.0', 'uniform']) <= set(ele)
#               ])  
    
    fig, axes = plt.subplots(nrows=max(len(rows),2), ncols=max(len(cols),2), figsize=(20, 17), sharex=False, sharey=False)
    for i, v1 in enumerate( rows ):
        if not use_discriminating:
            col_vals= our[ (our[cols_names[0]] == v1) ]['num_models'].dropna()
            ylim = (0, col_vals.max())
        else:
            col_vals = our[ (our[cols_names[0]] == v1) ]['discriminating'].dropna().max()
            ylim = (0, max(3, col_vals.max()))
        
        for j, v2 in enumerate( cols ):
            if pt[v1][v2].empty: # not done yet
                continue
    
            colss = tuple(sorted(pt[v1][v2].columns))
            cc = list(styles_mapping[colss])
            if len(cc) == 1:
                cc=cc[0]
#             display(list(styles_mapping[colss]))
            pt[v1][v2].dropna().plot(ax=axes[i][j],
                                     title="{}: {},   {}: {}".format(cols_names[1], v2, cols_names[0], v1), 
                                     color=cc, 
                                     marker='o',
                                     legend=legend,
                                     xlim=r_limits,
                                     ylim=ylim)
    
        axes[i][0].set_ylabel(x_axis_name)
        
#     fig.tight_layout()
    display(pt)
    
    fig.legend( axes[max_val[0]][max_val[1]].get_lines() ,
               legend_names, 
               loc = 'upper center', 
               ncol=5 )
```
