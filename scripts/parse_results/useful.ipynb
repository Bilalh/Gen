```python
def plot_grid(*, our, cols_names, use_discriminating=False, legend=True ):
    
    if use_discriminating:
        values='discriminating'
        x_axis_name='number of distinguishing params'
        pass
    else:
        values='resulting_models_filled'
        x_axis_name='models left'
    
    pt=our.pivot_table(rows=['total_timeout_h'],
                       cols=cols_names,
                       values=values)
    
    rows = our[cols_names[0]].unique()
    cols = our[cols_names[1]].unique()
    print("rows:", rows, "cols:", cols)
    
    r_limits =( df['total_timeout_h'].dropna().min(),  df['total_timeout_h'].dropna().max())
    
    lens = ( (i,j, pt[r][c].columns ) for i, r in  enumerate(rows)  for j, c in enumerate(cols) )
    max_val = max(lens,key=lambda v: len(v[2]))
    *max_val, legend_names = max_val
    
    styles = ['bo-','ro-','go-', 'yo-', 'mo-', 'ko-']
    styles_mapping = make_style_mapping(legend_names, styles)
    
    fig, axes = plt.subplots(nrows=max(len(rows),2), ncols=max(len(cols),2), figsize=(22, 17), sharex=True, sharey='row')
    for i, v1 in enumerate( rows ):
        if not use_discriminating:
            col_vals= our[ (our[cols_names[0]] == v1) ]['num_models'].dropna()
            ylim = (0, col_vals.max())
        else:
            col_vals = our[ (our[cols_names[0]] == v1) ]['discriminating'].dropna().max()
            ylim = (0, max(3, col_vals.max()))
        
        for j, v2 in enumerate( cols ):
            if pt[v1][v2].empty: # not done yet
                continue
    
            col_names = tuple(pt[v1][v2].columns)
            pt[v1][v2].dropna().plot(ax=axes[i][j],
                                     title="{}: {},   {}: {}".format(cols_names[1], v2, cols_names[0], v1), 
                                     style=list(styles_mapping[col_names]), 
                                     legend=legend,
                                     xlim=r_limits,
                                     ylim=ylim)
    
        axes[i][0].set_ylabel(x_axis_name)
        
    
    fig.legend( axes[max_val[0]][max_val[1]].get_lines() ,
               legend_names, 
               loc = 'upper center', 
               ncol=5 )
    
    return pt
```

```python
our = df [ (df['method'] == 'ksample')  ]

pt = our[ (our['total_timeout_h'].isin( [6,12,24]) )  ].pivot_table(
                    rows=['run_no'],
                    cols=['essence', 'total_timeout_h', 'races', 'num_points', 'point_selector'],
                    values='resulting_models_filled')
# display(pt)
pt_var = pt.var()
ll = pt_var[ (pt_var < 1000) ]

low_var=pt.filter(items=ll.index)

by_common = low_var.mean().unstack(level='essence').unstack(level='total_timeout_h').unstack(level='races')
out=by_common.stack().stack().stack()

mins = by_common.min().dropna().unstack(['essence','races'])
# display(mins)

def unnest(t):
    return t[0] + (t[1],)
items = [ list(zip(out.index.names + ['resulting_models_filled'], unnest(t))) for t in  out.items() ]

results_k = []
for item in items:
    d = dict(item)
    min_left = mins[d['essence']][ d['races'] ][d['total_timeout_h']]
    if d['resulting_models_filled'] <= (min_left + 0.0001) :
        res = list(zip(*item))[1]
        results_k.append(res)
    

ksample = pd.DataFrame.from_records( results_k, 
                                    columns=out.index.names + ['resulting_models_filled'],
                                    coerce_float = True)

ksample
# ksample.pivot_table(
#                     rows=['total_timeout_h'],
#                     cols=['essence', 'races'],
#                     values='resulting_models_filled')
```
