```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

from analyse import *

df = load_data_frame()
pd.options.display.line_width = 110
pd.options.display.max_columns = 50
```

```python
df
```

```python
our = df[ (df['total_timeout_h'] >=1 ) ]

pt=our.pivot_table(rows=['total_timeout_h'],
                   cols=['essence','races','method' ],
                   values='resulting_models_filled')

cols = our['races'].unique()
rows = our['essence'].unique()

lens = ( (i,j, pt[r][c].columns ) for i, r in  enumerate(rows)  for j, c in enumerate(cols) )
max_val = max(lens,key=lambda v: len(v[2]))
*max_val, legend_names = max_val

styles = ['bo-','ro-','go-', 'yo-', 'mo-', 'ko-']
styles_mapping = make_style_mapping(legend_names, styles)


fig, axes = plt.subplots(nrows=len(rows), ncols=len(cols), figsize=(22, 17), sharex=True, sharey='row')
for i, v1 in enumerate( rows ):
    for j, v2 in enumerate( cols ):
        col_names = tuple(pt[v1][v2].columns)
        pt[v1][v2].dropna().plot(ax=axes[i][j],
                                 title="{} races, {}".format(v2, v1), 
                                 style=list(styles_mapping[col_names]), 
                                 legend=False)
    axes[i][0].set_ylabel('models left')
    

fig.legend( axes[max_val[0]][max_val[1]].get_lines() ,
           legend_names, 
           loc = 'upper center', 
           ncol=5 )

from matplotlib.backends.backend_pdf import PdfPages
with PdfPages('/Users/bilalh/Desktop/a.pdf') as pdf:
    pdf.savefig()
    
# fig.savefig('/Users/bilalh/Desktop/a.png')
```
