```python
def plot_grid(*, our, cols_names, use_discriminating=False, legend=True ):
    
    if use_discriminating:
        values='discriminating'
        x_axis_name='number of distinguishing params'
        pass
    else:
        values='resulting_models_filled'
        x_axis_name='models left'
    
    pt=our.pivot_table(rows=['total_timeout_h'],
                       cols=cols_names,
                       values=values)
    
    rows = our[cols_names[0]].unique()
    cols = our[cols_names[1]].unique()
    print("rows:", rows, "cols:", cols)
    
    r_limits =( df['total_timeout_h'].dropna().min(),  df['total_timeout_h'].dropna().max())
    
    lens = ( (i,j, pt[r][c].columns ) for i, r in  enumerate(rows)  for j, c in enumerate(cols) )
    max_val = max(lens,key=lambda v: len(v[2]))
    *max_val, legend_names = max_val
    
    styles = ['bo-','ro-','go-', 'yo-', 'mo-', 'ko-']
    styles_mapping = make_style_mapping(legend_names, styles)
    
    fig, axes = plt.subplots(nrows=max(len(rows),2), ncols=max(len(cols),2), figsize=(22, 17), sharex=True, sharey='row')
    for i, v1 in enumerate( rows ):
        if not use_discriminating:
            col_vals= our[ (our[cols_names[0]] == v1) ]['num_models'].dropna()
            ylim = (0, col_vals.max())
        else:
            col_vals = our[ (our[cols_names[0]] == v1) ]['discriminating'].dropna().max()
            ylim = (0, max(3, col_vals.max()))
        
        for j, v2 in enumerate( cols ):
            if pt[v1][v2].empty: # not done yet
                continue
    
            col_names = tuple(pt[v1][v2].columns)
            pt[v1][v2].dropna().plot(ax=axes[i][j],
                                     title="{}: {},   {}: {}".format(cols_names[1], v2, cols_names[0], v1), 
                                     style=list(styles_mapping[col_names]), 
                                     legend=legend,
                                     xlim=r_limits,
                                     ylim=ylim)
    
        axes[i][0].set_ylabel(x_axis_name)
        
    
    fig.legend( axes[max_val[0]][max_val[1]].get_lines() ,
               legend_names, 
               loc = 'upper center', 
               ncol=5 )
    
    return pt
```

```python
our = df [ (df['method'] == 'ksample')  ]

pt = our[ (our['total_timeout_h'].isin([6,12,24]) )  ].pivot_table(
                    rows=['run_no'],
                    cols=['essence', 'total_timeout_h', 'races', 'num_points', 'point_selector'],
                    values='resulting_models_filled')
# display(pt)
pt_var = pt.var()
ll = pt_var[ (pt_var < 1000) ]

low_var=pt.filter(items=ll.index)

by_common = low_var.mean().unstack(level='essence').unstack(level='total_timeout_h').unstack(level='races')
out=by_common.stack().stack().stack()

#TODO best overall
# display(low_var.mean().unstack(level=['point_selector', 'num_points' ]  ))

mins = by_common.min().dropna().unstack(['essence','races'])
display(mins)

col_names = out.index.names + ['resulting_models_filled', 'num_models', 'method']

def unnest(t):
    return t[0] + (t[1],)
items = [ list(zip(col_names, unnest(t))) for t in  out.items() ]

results_k = []
for item in items:
    d = dict(item)
    min_left = mins[d['essence']][ d['races'] ][d['total_timeout_h']]
    if d['resulting_models_filled'] <= (min_left + 0.0001) :
        res = list(zip(*item))[1]
        res += (df[ df['essence'] == d['essence'] ]['num_models'][0:1], 'ksample' )
        results_k.append(res)
    

ksample = pd.DataFrame.from_records(results_k, 
                                    columns=col_names,
                                    coerce_float = True)

ksample.pivot_table(
                    rows=['total_timeout_h'],
                    cols=['essence','races', 'num_points', 'point_selector'],
                    values='resulting_models_filled')
```

```python
def row_for_display(row):
    to_drop =['discriminating', 'output_dir', 'resulting_models_filled','total_timeout']
    for col in ['quality', 'resulting_models', 'radius_as_percentage']:
        if col in row:
            to_drop.append(col)

    to_drop += ['essence','total_timeout_h', 'races','models_timeout', 'group_num','run_no']
    
    to_display=row.drop(labels=to_drop).dropna()
    
#     show_first =[ 'essence', 'num_models', 'method', 'total_timeout_h',  'races', 'run_no']
    show_first =['method', 'num_models']
    s1 = str(to_display[show_first]).splitlines()[0:-1]
    s2 = str(to_display.drop(labels=show_first)).splitlines()[0:-1]
    
    def check(f,s):
        if f:
            return f + "    " + s
        else:
            return s
    result = "\n".join( check(f,s)  for (f, s) in it.zip_longest(s1,s2, fillvalue="") ) 
    return result
        
```

```python
param_eprime_results_sql="""

Select count(param) as 'param#',eprimes_num as 'eprime#',  eprimes
From (
    Select P.param, count(D.eprime) as eprimes_num  , group_concat(D.eprime, ", ") as eprimes, quality
    From ParamQuality P
    Join TimingsDomination D on P.param = D.param
    Where P.Quality < 1 and isDominated = 0
    Group by P.param

	Union

	Select Q.param, 0 as eprimes_num, "" as eprimes, 100 as quality
	From ParamQuality Q
	Where Q.param not in (Select D.param From DiscriminatingParams D)

    Order by quality

) 

Group by eprimes
Order by eprimes_num,count(param), eprimes;
"""


import sqlite3
def get_models_for_all_params_db(row):
    db_path = Path(row['output_dir'])  / "results.db"
    eprimes = pd.read_sql(sql=param_eprime_results_sql, 
                         con=sqlite3.Connection(str(db_path)),
                         )
    return eprimes
    
def get_models_for_all_params(row):
#     db_path = Path(out_dir)  / "results.db"
#     eprimes = pd.read_sql(sql=param_eprime_results_sql, 
#                          con=sqlite3.Connection(str(db_path)),
#                          )
    eprimes = param_eprime_info[ param_eprime_info["index"] == row["index"] ]
    return eprimes[[ 'param#', 'eprime#', 'eprimes' ]]

p="/Users/bilalh/Desktop/Experiments/results/ksample/prob024-Langford/out-43200-10-32__1___10_True_100_halving"
index = df[ df['output_dir'] == Path(p) ]['index'].values[0]
t_db=get_models_for_all_params_db({"output_dir":p})
t_cached=get_models_for_all_params({'index':47})
display(t_db, t_cached)
```
